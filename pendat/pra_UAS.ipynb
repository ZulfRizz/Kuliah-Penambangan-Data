# Preprocessing Data

Tahap Preprocessing atau Data Preparation adalah salah satu langkah paling krusial dalam siklus machine learning. Tujuannya adalah membersihkan dan mengubah data mentah menjadi format yang bersih, terstruktur, dan sesuai untuk dimasukkan ke dalam model. Tanpa pra-pemrosesan yang baik, performa model bisa menjadi tidak akurat atau bias.

Pada tahap ini, kita akan melakukan dua kegiatan utama: Transformasi Data dan Normalisasi/Skala.

```python
# Asumsi 'df' adalah DataFrame yang sudah bersih dari duplikasi dari tahap sebelumnya
# Salin dataframe untuk menjaga data asli
df_processed = df.copy()
```

## Transformasi Data

### A. Membuat Variabel Target (Dependent Variable)
Tujuan kita adalah klasifikasi biner ("Lulus" / "Gagal"). Kita perlu membuat kolom target ini dari kolom G3 yang bersifat numerik. Sesuai definisi sebelumnya, siswa dianggap "Lulus" jika G3 >= 10.

```python
# Membuat kolom 'status' berdasarkan nilai G3
# 'Lulus' jika G3 >= 10, sebaliknya 'Gagal'
df_processed['status'] = np.where(df_processed['G3'] >= 10, 'Lulus', 'Gagal')

# Mengubah target menjadi numerik untuk beberapa model (1 untuk Lulus, 0 untuk Gagal)
df_processed['status_binary'] = df_processed['status'].map(
    {'Lulus': 1, 'Gagal': 0})


# Menampilkan hasil transformasi
print("Contoh hasil pembuatan kolom target:")
display(df_processed[['G3', 'status', 'status_binary']].head())

# Melihat distribusi target
print("\nDistribusi kelas target:")
print(df_processed['status'].value_counts())
```

### B. Mengubah Fitur Kategorikal menjadi Numerik (Encoding)
Model machine learning tidak dapat memproses data teks. Oleh karena itu, semua fitur kategorikal harus diubah menjadi angka. Ada dua pendekatan utama yang akan kita gunakan:

- Label Encoding/Mapping: Untuk fitur biner (hanya memiliki 2 kategori), kita bisa memetakannya ke 0 dan 1.
- One-Hot Encoding: Untuk fitur nominal (memiliki >2 kategori tanpa urutan), kita membuat kolom baru untuk setiap kategori.


```python
# --- 1. Encoding Fitur Biner (Mapping) ---
# Daftar fitur biner 'yes'/'no'
binary_map_cols = ['schoolsup', 'famsup', 'paid',
                    'activities', 'nursery', 'higher', 'internet', 'romantic']
for col in binary_map_cols:
    df_processed[col] = df_processed[col].map({'yes': 1, 'no': 0})

# Mapping manual untuk fitur biner lainnya
df_processed['sex'] = df_processed['sex'].map({'F': 1, 'M': 0})
df_processed['address'] = df_processed['address'].map({'U': 1, 'R': 0})
df_processed['famsize'] = df_processed['famsize'].map({'GT3': 1, 'LE3': 0})
df_processed['Pstatus'] = df_processed['Pstatus'].map({'T': 1, 'A': 0})
df_processed['school'] = df_processed['school'].map({'GP': 1, 'MS': 0})


# --- 2. One-Hot Encoding untuk Fitur Nominal ---
# Daftar fitur nominal dengan lebih dari 2 kategori
nominal_cols = ['Mjob', 'Fjob', 'reason', 'guardian']

# Menerapkan one-hot encoding
df_processed = pd.get_dummies(
    df_processed, columns=nominal_cols, drop_first=True)

print("Dataset setelah encoding:")
display(df_processed.head())
```

### C. Memilih Fitur dan Target Final (Feature Selection)
Sekarang kita akan memisahkan dataset menjadi himpunan fitur (X) dan target (y). Pada langkah ini, kita juga akan membuang kolom-kolom yang tidak diperlukan lagi.

- `G1`, `G2`, `G3`: Dibuang karena sangat berkorelasi dengan target dan menyebabkan kebocoran data (data leakage). Kita ingin memprediksi hasil akhir sebelum nilai-nilai ini diketahui.
  
- `status`: Kolom target versi teks, kita akan menggunakan `status_binary`.

```python
# Memisahkan fitur (X) dan target (y)
X = df_processed.drop(columns=['G1', 'G2', 'G3', 'status', 'status_binary'],inplace=False)
y = df_processed['status_binary']

print("Bentuk matriks fitur (X):", X.shape)
print("Bentuk vektor target (y):", y.shape)
```

## Normalisasi

Beberapa algoritma machine learning (seperti SVM, Regresi Logistik, KNN) sangat sensitif terhadap skala fitur. Fitur dengan rentang nilai yang besar (misalnya absences) dapat mendominasi fitur dengan rentang kecil (misalnya studytime), yang dapat mengganggu proses pembelajaran model.

Untuk mengatasi ini, kita perlu menyamakan skala semua fitur numerik. Kita akan menggunakan `StandardScaler`

```python
from sklearn.preprocessing import StandardScaler

# Kita akan membuat salinan baru dari X untuk perbandingan
X_standard_scaled = X.copy()

# Mengidentifikasi kolom numerik yang perlu diskalakan (sama seperti sebelumnya)
numerical_cols_to_scale = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures',
                            'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']

# Inisialisasi scaler
standard_scaler = StandardScaler()

# Terapkan scaler HANYA pada kolom numerik yang telah diidentifikasi
X_standard_scaled[numerical_cols_to_scale] = standard_scaler.fit_transform(
    X_standard_scaled[numerical_cols_to_scale])

print("Lima baris pertama dari data fitur (X) setelah standardisasi Z-Score:")
display(X_standard_scaled.head())

print("\nRingkasan statistik setelah standardisasi (perhatikan nilai 'mean' dan 'std'):")
# Tampilkan statistik untuk verifikasi
# Mean akan sangat mendekati 0, dan Std Dev akan sangat mendekati 1
display(X_standard_scaled[numerical_cols_to_scale].describe().transpose())
```

# Modeling

Pada tahap ini, kita akan menggunakan data yang telah diproses (X dan y) untuk melatih model machine learning. Prosesnya meliputi:

- Membagi data menjadi set pelatihan dan pengujian.
- Melatih model Naive Bayes dan Decision Tree pada set pelatihan.
- Mengevaluasi performa masing-masing model pada set pengujian.
- Membandingkan hasil kedua model.

```python
# Import library yang diperlukan untuk modeling dan evaluasi
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Kita akan menggunakan data yang telah distandarisasi dengan Z-Score
# Jika Anda lebih memilih RobustScaler, ganti X_standard_scaled menjadi X_robust_scaled
# Asumsi X_standard_scaled dan y sudah ada dari tahap sebelumnya
X_final = X_standard_scaled
```

## Pembagian Data (Train-Test Split)
Langkah pertama dan paling penting sebelum melatih model adalah membagi dataset menjadi dua bagian:

- Training Set: Data yang digunakan untuk "mengajari" atau melatih model (umumnya 80%).
  
- Testing Set: Data yang "disembunyikan" dari model selama pelatihan, digunakan untuk menguji seberapa baik performa model pada data baru yang belum pernah dilihatnya (umumnya 20%).
Ini penting untuk mencegah overfitting, di mana model hanya hafal data latihan tetapi tidak bisa menggeneralisasi dengan baik.

```python
# Membagi data dengan perbandingan 80% train dan 20% test
# stratify=y memastikan proporsi kelas 'Lulus' dan 'Gagal' sama di train dan test set
# random_state=42 memastikan hasil pembagian data selalu sama setiap kali kode dijalankan
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42, stratify=y)

print("Bentuk data Latih (X_train):", X_train.shape)
print("Bentuk data Uji (X_test):", X_test.shape)
print("Bentuk target Latih (y_train):", y_train.shape)
print("Bentuk target Uji (y_test):", y_test.shape)
```

## Model 1: Gaussian Naive Bayes
Naive Bayes adalah algoritma klasifikasi probabilistik berdasarkan Teorema Bayes dengan asumsi "naive" (naif) bahwa semua fitur bersifat independen satu sama lain. Gaussian Naive Bayes adalah varian yang digunakan ketika fitur-fitur prediktornya bersifat kontinu dan diasumsikan terdistribusi secara normal (Gaussian), yang cocok dengan data kita yang telah distandarisasi.

```python
# 1. Inisialisasi Model
nb_model = GaussianNB()

# 2. Latih Model
nb_model.fit(X_train, y_train)

# 3. Lakukan Prediksi
y_pred_nb = nb_model.predict(X_test)

# 4. Evaluasi Model
print("--- Hasil Evaluasi Model Naive Bayes ---")
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print(f"Akurasi: {accuracy_nb:.4f}")
print("\nLaporan Klasifikasi:")
print(classification_report(y_test, y_pred_nb,
      target_names=['Gagal (0)', 'Lulus (1)']))

# Visualisasi Confusion Matrix
cm_nb = confusion_matrix(y_test, y_pred_nb)
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Blues', xticklabels=[
            'Gagal', 'Lulus'], yticklabels=['Gagal', 'Lulus'])
plt.title('Confusion Matrix - Naive Bayes')
plt.ylabel('Kelas Aktual')
plt.xlabel('Kelas Prediksi')
plt.show()
```

## Model 2: Decision Tree Classifier
Decision Tree adalah model non-parametrik yang membuat keputusan dengan cara memecah data menjadi subset yang lebih kecil berdasarkan serangkaian aturan "if-then-else". Model ini sangat intuitif karena strukturnya mirip dengan diagram alir (flowchart).

```python
# 1. Inisialisasi Model
# random_state=42 untuk reproduktifitas
dt_model = DecisionTreeClassifier(random_state=42)

# 2. Latih Model
dt_model.fit(X_train, y_train)

# 3. Lakukan Prediksi
y_pred_dt = dt_model.predict(X_test)

# 4. Evaluasi Model
print("--- Hasil Evaluasi Model Decision Tree ---")
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Akurasi: {accuracy_dt:.4f}")
print("\nLaporan Klasifikasi:")
print(classification_report(y_test, y_pred_dt,
      target_names=['Gagal (0)', 'Lulus (1)']))

# Visualisasi Confusion Matrix
cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', xticklabels=[
            'Gagal', 'Lulus'], yticklabels=['Gagal', 'Lulus'])
plt.title('Confusion Matrix - Decision Tree')
plt.ylabel('Kelas Aktual')
plt.xlabel('Kelas Prediksi')
plt.show()
```

## Evaluasi dan Perbandingan Model

```python
# --- Impor Library yang Dibutuhkan ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    roc_auc_score
)

# Asumsi variabel berikut sudah ada dari tahap sebelumnya:
# y_test: Nilai target sebenarnya dari set pengujian
# y_pred_nb: Prediksi dari model Naive Bayes
# y_pred_dt: Prediksi dari model Decision Tree
# nb_model: Model Naive Bayes yang sudah dilatih
# dt_model: Model Decision Tree yang sudah dilatih
# X_test: Data fitur dari set pengujian

# --- 1. Laporan Klasifikasi Rinci ---

print("="*60)
print("LAPORAN EVALUASI MODEL GAUSSIAN NAIVE BAYES")
print("="*60)
print(classification_report(y_test, y_pred_nb,
      target_names=['Gagal (0)', 'Lulus (1)']))

print("\n" + "="*60)
print("LAPORAN EVALUASI MODEL DECISION TREE")
print("="*60)
print(classification_report(y_test, y_pred_dt,
      target_names=['Gagal (0)', 'Lulus (1)']))


# --- 2. Visualisasi Confusion Matrix Berdampingan ---

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Perbandingan Confusion Matrix', fontsize=16)

# Confusion Matrix untuk Naive Bayes
cm_nb = confusion_matrix(y_test, y_pred_nb)
sns.heatmap(cm_nb, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Gagal', 'Lulus'], yticklabels=['Gagal', 'Lulus'])
axes[0].set_title('Naive Bayes')
axes[0].set_ylabel('Kelas Aktual')
axes[0].set_xlabel('Kelas Prediksi')

# Confusion Matrix untuk Decision Tree
cm_dt = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm_dt, annot=True, fmt='d', cmap='Greens', ax=axes[1],
            xticklabels=['Gagal', 'Lulus'], yticklabels=['Gagal', 'Lulus'])
axes[1].set_title('Decision Tree')
axes[1].set_ylabel('Kelas Aktual')
axes[1].set_xlabel('Kelas Prediksi')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


# --- 3. Tabel Perbandingan Metrik Utama ---

# Mengumpulkan metrik untuk kedua model
metrics = {
    'Model': ['Gaussian Naive Bayes', 'Decision Tree'],
    'Akurasi': [
        accuracy_score(y_test, y_pred_nb),
        accuracy_score(y_test, y_pred_dt)
    ],
    "Presisi (Gagal)": [
        precision_score(y_test, y_pred_nb, pos_label=0),
        precision_score(y_test, y_pred_dt, pos_label=0)
    ],
    "Recall (Gagal)": [
        recall_score(y_test, y_pred_nb, pos_label=0),
        recall_score(y_test, y_pred_dt, pos_label=0)
    ],
    "F1-Score (Gagal)": [
        f1_score(y_test, y_pred_nb, pos_label=0),
        f1_score(y_test, y_pred_dt, pos_label=0)
    ],
    "F1-Score (Lulus)": [
        f1_score(y_test, y_pred_nb, pos_label=1),
        f1_score(y_test, y_pred_dt, pos_label=1)
    ]
}

# Membuat DataFrame dari dictionary
comparison_df = pd.DataFrame(metrics)
comparison_df = comparison_df.set_index('Model')

print("\n" + "="*60)
print("TABEL PERBANDINGAN METRIK MODEL")
print("="*60)
display(comparison_df.style.format(
    "{:.4f}").background_gradient(cmap='viridis_r'))


# --- 4. Kurva ROC (Receiver Operating Characteristic) dan Skor AUC ---

# Dapatkan probabilitas prediksi untuk kelas positif (Lulus = 1)
y_prob_nb = nb_model.predict_proba(X_test)[:, 1]
y_prob_dt = dt_model.predict_proba(X_test)[:, 1]

# Hitung FPR, TPR, dan AUC
fpr_nb, tpr_nb, _ = roc_curve(y_test, y_prob_nb)
auc_nb = roc_auc_score(y_test, y_prob_nb)

fpr_dt, tpr_dt, _ = roc_curve(y_test, y_prob_dt)
auc_dt = roc_auc_score(y_test, y_prob_dt)

# Plot Kurva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr_nb, tpr_nb,
         label=f'Naive Bayes (AUC = {auc_nb:.4f})', color='blue')
plt.plot(fpr_dt, tpr_dt,
         label=f'Decision Tree (AUC = {auc_dt:.4f})', color='green')
plt.plot([0, 1], [0, 1], 'k--', label='Garis Acak')  # Garis referensi
plt.title('Kurva ROC (Receiver Operating Characteristic)')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend
```
